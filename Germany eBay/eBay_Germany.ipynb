##### Some observations

- Regn yr seem to have a value = 1000 and 9999
- Odometer, power_ps, regn yr, price values should be converted to integer/float

```python
autos.loc[4:12, "price"]
```

 - Lets remove the $ symbol and comma from price and make it a float

```python
autos["price"] = autos["price"].str.replace("$", "")
autos["price"] = autos["price"].str.replace(",", "")
autos["price"] = autos["price"].astype(int)
```

- Lets do the same for odometer

```python
autos["odometer"] = autos["odometer"].str.replace("km", "")
autos["odometer"] = autos["odometer"].str.replace(",", "")
autos["odometer"] = autos["odometer"].astype(int)
```

--- Volkswagen has around 21% entries and Suzuki has around 6% entries. Let's choose all those with greater than 5% entries


```python
brands_above_5_pct = autos["brand"].value_counts(normalize=True).index[0:27]
brands_above_5_pct
```

```python
mean_price_by_brand = {}
for brand in brands_above_5_pct:
    mean_price = autos[autos["brand"] == brand]["price"].mean()
    mean_price_by_brand[brand] = mean_price
    
mean_price_by_brand
```

---Now we have the top brands with their mean prices aggregated for further analysis and inference.

```python
mean_mileage_by_brand = {}
for brand in brands_above_5_pct:
    mean_mileage = autos[autos["brand"] == brand]["odometer_km"].mean()
    mean_mileage_by_brand[brand] = mean_mileage
    
mean_mileage_by_brand
```

```python
bmp_series = pd.Series(mean_price_by_brand)
print(bmp_series)
bmm_series = pd.Series(mean_mileage_by_brand)
print(bmm_series)
```

```python
df = pd.DataFrame(bmp_series, columns=['mean_price'])
```

```python
df["mean_mileage"] = bmm_series
```
